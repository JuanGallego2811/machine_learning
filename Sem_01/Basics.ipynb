```
# Cargar la imagen
img = cv2.imread("/content/drive/MyDrive/Machine learning/machine_learning/Images/lena.jpg")

# Convertir de BGR a RGB (porque OpenCV carga la imagen en formato BGR por defecto)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tamaño del cuadro
cuadro_size = 50

# Colores base para los cuadros (RGB)
colores_base = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]

# Iterar sobre los cuadros y aplicar colores intercalados
for i in range(0, img.shape[0], cuadro_size):
    for j in range(0, img.shape[1], cuadro_size):
        # Calcular el índice del color base según la posición del cuadro
        indice_color = (i -1 // cuadro_size + j // cuadro_size) % len(colores_base)
        # Obtener el color base correspondiente
        color_base = colores_base[indice_color]
        # Asignar el color al canal correspondiente en el cuadro
        img[i:i+cuadro_size, j:j+cuadro_size, indice_color] = color_base[indice_color ]

# Mostrar la imagen con el patrón de cuadros generados
plt.imshow(img)
plt.show()
```

```
# Cargar la imagen
img = cv2.imread("/content/drive/MyDrive/Machine learning/machine_learning/Images/lena.jpg")

# Convertir de BGR a RGB (porque OpenCV carga la imagen en formato BGR por defecto)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tamaño del cuadro
cuadro_size = 50

# Colores base para los cuadros (RGB)
colores_base = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]

# Iterar sobre los cuadros y aplicar colores intercalados
for i in range(0, img.shape[0], cuadro_size):
    for j in range(0, img.shape[1], cuadro_size):
        # Calcular el índice del color base según la posición del cuadro
        indice_color = (i // cuadro_size + j // cuadro_size) % len(colores_base)
        # Obtener el color base correspondiente
        color_base = colores_base[indice_color]
        # Asignar el color al cuadro
        img[i:i+cuadro_size, j:j+cuadro_size] = color_base

# Mostrar la imagen con el patrón de cuadros generados
plt.imshow(img)
plt.show()
```
