# Mean Shift - Generated data

Mean shift documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.MeanShift.html)

```python
import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
n_samples = 1000  # Coloca aquí el número deseado de muestras
random_state = 3  # Coloca aquí el estado aleatorio deseado

X, _ = make_blobs(n_samples=n_samples, random_state=random_state)
X_train, X_test = train_test_split(X, test_size=0.2)
_, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].scatter(X[:, 0], X[:, 1], c="b", edgecolors='k')
axes[1].scatter(X_train[:, 0], X_train[:, 1], c="r", edgecolors='k')
axes[1].scatter(X_test[:, 0], X_test[:, 1], c="g", edgecolors='k')
```

```python
bandwidth_h = max(abs(X[:,0].min() - X[:,0].max()) , abs(X[:,1].min() - X[:,1].max()))*0.1
print(bandwidth_h)
```

```python
ms = MeanShift(bandwidth = bandwidth_h)
ms.fit(X_train)

cluster_centers = ms.cluster_centers_
```

```python
predictions = ms.predict(X_test)

_,ax = plt.subplots(figsize = (6,4))
ax.scatter(X_test[:,0], X_test[:,1] ,c = predictions, cmap = 'Paired',edgecolors = 'k')
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], c='m', s = 250,edgecolors = 'k' , alpha = 0.8 , marker = "*")
```
